```python
%matplotlib inline

from matplotlib.animation import FuncAnimation
from IPython.display import HTML

postprocess_fn=partial(data.postprocess, num_bits=config.data.num_bits)

def frames_to_anim(frames):
    plt.figure()
    patch = plt.imshow(frames[0])
    plt.axis('off')

    def animate(i):
        patch.set_data(frames[i])

    anim = FuncAnimation(plt.gcf(), animate, frames = len(frames), interval=50)
    plt.close()
    return anim

def interpolate(y1,y2):

    num_samples = 100
    sampling_shape = (config.data.num_channels*config.data.image_size**2,)
    key = jax.random.PRNGKey(0)

    l = np.linspace(0,1,num_samples).reshape(-1,1)
    y = l*y1[None] + (1-l)*y2[None]

    x, _ = model.apply(params,y,reverse=True)
    frames = np.squeeze(postprocess_fn(x))
    
    return frames
```

```python
x,y = next(val_ds)

frames = interpolate(y[0],y[1])
```

```python
HTML(frames_to_anim(frames).to_jshtml())
```
